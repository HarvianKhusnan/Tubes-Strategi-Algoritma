# Greedy


## Greed By Nilai
Misal kita memilik modal Rp 1000.000 untuk membeli saham. Agar lebih efektif, kita tidak menghabiskan seluruh modal kita untuk satu jenis saham. Maka dari itu kita memutuskan untuk membeli satu jenis saham dengan max 100 lembar.

```python
maxNilai = saham.sort_values(["Nilai"], ascending=False)
maxNilai = maxNilai.reset_index()
maxNilai = maxNilai.drop('index', axis=1)
maxNilai
```

```python
def sahamGBN(maxNilai,modal):
  total = 0
  dibeli = []
  i = 0
  while i < (len(maxNilai)-1) and total < modal:
    for j in range(100):
      if total + maxNilai.Sebelumnya[i] < modal or total + maxNilai.Sebelumnya[i] == modal:
        total += maxNilai.Sebelumnya[i]
      else:
        break
    if j != 0:
      dibeli += [[maxNilai.Nama_Perusahaan[i],j+1]]
    i += 1
  
  for i in range(len(dibeli)):
    print(dibeli[i])
  print("Total : ",total)

sahamGBN(maxNilai,1000000)
# maxNilai
```
